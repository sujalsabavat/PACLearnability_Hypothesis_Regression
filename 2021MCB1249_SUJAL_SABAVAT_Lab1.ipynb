```python
from numpy import *

def PartitionRaja_Discover(X, Y, k):

    assert all(array(X) == array(sorted(X))), "X must be sorted!"

    X = array(X)
    Y = array(Y)
    m = len(X)
    DP1 = [[None for j in range(k+1)] for i in range(m+1)]
    DP2 = zeros((m+1, k+1), dtype=int)

    cuml_sum_y = concatenate([[0], cumsum(Y)])

    DP2[:m+1,0] = cuml_sum_y

    for i in range(1, m+1):
        for j in range(1, k+1):
            choices = []
            for l in range(0,i+1):
                nxt_errs = DP2[l,j-1] + (cuml_sum_y[i]-cuml_sum_y[l]) + concatenate([[0], cumsum((-1)**(Y[arange(l, i)] == 1))])
                min_err = argmin(nxt_errs)
                choices.append((nxt_errs[min_err], (l, arange(l,i+1)[min_err])))

            DP2[i,j], DP1[i][j] = min(choices)

    good = []
    cur = DP1[m][k]
    for i in range(k,0,-1):
        good.append(cur)
        cur = DP1[cur[0]][i-1]
        if cur == None:
            break
    good = sorted(good)
    temp = concatenate([[0], X, [1]])
    reps = (temp[1:]+temp[:-1]) / 2.0
    partitions = [(reps[a], reps[b]) for a,b in good]

    return partitions
```

```python
def find_error(Y_true,Y_pred):
    import numpy as np

    n_samples = len(Y_true)
    return np.mean((Y_true - Y_pred) ** 2) / n_samples
    pass


def f(x, z):
    import numpy as np
    return -2 * np.cos(x) / np.pi + (2 * x) / (2 * np.pi) - 2 * np.cos(3 * x) / (3 * np.pi) + z * np.random.normal(loc=0, scale=1)
```

```python
def generate_data_with_gaussian_noise(scaling_factor, total_number_of_data_points):
    import numpy as np


    X = np.random.uniform(low=-10, high=10, size=total_number_of_data_points)
    X = (X - np.min(X)) / (np.max(X) - np.min(X))
    y = [custom_function(x, scaling_factor) for x in X]
    y = (y - np.min(y)) / np.std(y)
    return X, y

def custom_function(x, scaling_factor):
    import numpy as np
    return x ** 2 + scaling_factor * np.random.normal()

def polynomial_features(X, degrees):
    import numpy as np
    X_poly = np.column_stack([X**i for i in range(1, degrees + 1)])
    return X_poly

def initialize_weights(number_of_features):
    import numpy as np
    return np.zeros(number_of_features, dtype=float)

def gradient_descent(X, y, weight, learningrate, beta, max_iterations=10000):
    import numpy as np
    momentum = np.zeros_like(weight, dtype=float)
    n_samples = len(y)

    for _ in range(max_iterations):
        predictions = np.dot(X, weight)
        error = ((predictions - y) ** 2).astype(float)

        normalized_error = error / n_samples  #normalising the error obtained


        gradient = (2 * X.T.dot(predictions - y)) / n_samples
        momentum = beta * momentum + (1 - beta) * gradient
        weight -= learningrate * momentum

    return weight

```

```python
def polynomial_regressor_gaus(X, y, degrees, learningrate=1e-4, beta=0.95):

  X_polynomial = polynomial_features(X, degrees)
  weight = initialize_weights(X_polynomial.shape[1])
  weight = gradient_descent(X_polynomial, y, weight, learningrate, beta)

  return weight

```

```python
import numpy as np

def run_experiment_with_different_degrees_gaus(maximum_degree, scaling_factor, total_number_of_data_points):
  degrees = list(range(1, maximum_degree + 1))
  errors = []

  X, y = generate_data_with_gaussian_noise(scaling_factor, total_number_of_data_points)
  y_pred_array = []

  for degree in degrees:
    weight = polynomial_regressor_gaus(X, y, degree)
    X_polynomial = polynomial_features(X, degree)
    y_predict = np.dot(X_polynomial, weight)
    error = find_error(y, y_predict)
    errors.append(error)
    y_pred_array.append(y_predict)

  return degrees, errors, X, y, y_pred_array


def run_experiment_with_different_total_number_of_points_and_degree_gaus(maximum_degree, scaling_factor, min_points, max_points, step):

  total_points_range = range(min_points, max_points + 1, step)
  result = {}


  for total_points in total_points_range:
    degrees, errors, X, y, y_pred_array = run_experiment_with_different_degrees_gaus(maximum_degree, scaling_factor, total_points)


    result[total_points]  =   {'degrees': degrees, 'errors': errors, 'X': X, 'y': y, 'y_pred_array': y_pred_array}

  return result
  pass

```

```python
def f(x,z):

    return -2*np.cos(x)/np.pi + (2*x)/(2*np.pi)-2*np.cos(3*x)/3*np.pi +z*np.random.poisson(lam=1)


def generate_data_with_poisson_noise(scaling_factor, total_number_of_data_points):

    # Generating random samples for X
    X = np.random.uniform(low=-10, high=10, size=total_number_of_data_points)

    # Normalizing X to the range [0, 1]
    X = (X - np.min(X)) / (np.max(X) - np.min(X))

    # Generating target variable y using the function with scaling factor
    y = f(X, scaling_factor)

    # Normalizing y to have zero mean and unit standard deviation
    y = (y - np.min(y)) / np.std(y)

    return X, y

```

```python
def polynomial_features(X, degree):
    X_polynomial = np.column_stack([X**i for i in range(1, degree + 1)])
    return X_polynomial


def initialize_weights(num_features):
    return np.zeros(num_features, dtype=float)

def gradient_descent(X, y, weight, learning_rate, beta, max_iterations=10000):
    momentum = np.zeros_like(weight, dtype=float)
    n_samples = len(y)
    iterations = 0


    while iterations < max_iterations:
        predictions = np.dot(X, weight)
        error = ((predictions - y) ** 2).astype(float)
        normalized_error = error / n_samples  # Normalizing the error
        gradient = (2 * X.T.dot(predictions - y)) / n_samples
        momentum = beta * momentum + (1 - beta) * gradient
        weight -= learning_rate * momentum
        iterations += 1

    return weight
```

```python
def polynomial_regressor_poisson(X, y, degree, learningrate=1e-4, beta=0.95):

    X_polynomial = polynomial_features(X, degree)
    weight = initialize_weights(X_polynomial.shape[1])
    weight = gradient_descent(X_polynomial, y, weight, learningrate, beta)

    return weight
```

```python
def run_experiment_with_different_degrees_poisson(max_degree, scaling_factor, total_number_of_data_points):

    degrees = list(range(1, max_degree + 1))
    errors = []

    X, y = generate_data_with_poisson_noise(scaling_factor, total_number_of_data_points)
    y_pred_array = []
    iterations = 0

    while iterations < len(degrees):
        degree = degrees[iterations]
        weights = polynomial_regressor_poisson(X, y, degree)
        X_polynomial = polynomial_features(X, degree)
        y_predict = np.dot(X_polynomial, weights)
        error = find_error(y, y_predict)
        errors.append(error)
        y_pred_array.append(y_predict)
        iterations += 1

    return degrees, errors, X, y, y_pred_array

def run_experiment_with_different_total_number_of_points_and_degree_poisson(max_degree, scaling_factor, min_points, max_points, step):

    total_points_range = range(min_points, max_points + 1, step)
    results = {}
    iterations = 0

    while iterations < len(total_points_range):
        total_points = total_points_range[iterations]
        degrees, errors, X, y, y_pred_array = run_experiment_with_different_degrees_poisson(max_degree, scaling_factor, total_points)
        results[total_points] = {'degrees': degrees, 'errors': errors, 'X': X, 'y': y, 'y_pred_array': y_pred_array}
        iterations += 1

    return results
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_concept():
    p = np.random.uniform(0, 100)
    q = np.random.uniform(p, 100)
    r = np.random.uniform(0, 100)
    s = np.random.uniform(r, 100)
    return p, q, r, s
```

```python
import numpy as np
import matplotlib.pyplot as plt


def generate_dataset(concept, n_samples=100):
    p, q, r, s = concept
    x = np.random.uniform(0, 100, n_samples)
    y = np.random.uniform(0, 100, n_samples)
    labels = np.where((x >= p) & (x <= q) & (y >= r) & (y <= s), 1, 0)
    return x, y, labels

def generate_hypothesis_and_error(x, y, labels):
    positive_indices = np.where(labels == 1)[0]
    if len(positive_indices) > 0:
        h1_p = min(x[positive_indices])
        h1_q = max(x[positive_indices])
        h1_r = min(y[positive_indices])
        h1_s = max(y[positive_indices])
        h1_error = np.mean(np.abs(labels - np.where((x >= h1_p) & (x <= h1_q) & (y >= h1_r) & (y <= h1_s), 1, 0)))
        return (h1_p, h1_q, h1_r, h1_s), h1_error
    else:
        return None, None
```

```python
import numpy as np
import matplotlib.pyplot as plt


def calculate_error_on_new_data(x_new, y_new, labels_new, h_parameter):
    h_p, h_q, h_r, h_s = h_parameter
    error = np.mean(np.abs(labels_new - np.where((x_new >= h_p) & (x_new <= h_q) & (y_new >= h_r) & (y_new <= h_s), 1, 0)))
    return error
```

```python
import numpy as np
import matplotlib.pyplot as plt



def plot_data_and_hypotheses(x, y, labels, h_true, h1_parameter, h2_parameter=None):
    plt.scatter(x[labels == 1], y[labels == 1], color='blue', label='Positive Examples')
    plt.scatter(x[labels == 0], y[labels == 0], color='red', label='Negative Examples')

    h_true_rectangle = plt.Rectangle((h_true[0], h_true[2]), h_true[1]-h_true[0], h_true[3]-h_true[2], fill=None, edgecolor='green', linestyle='-.', label='True Concept')
    plt.gca().add_patch(h_true_rectangle)

    if h1_parameter:
        h1_rectangle = plt.Rectangle((h1_parameter[0], h1_parameter[2]), h1_parameter[1]-h1_parameter[0], h1_parameter[3]-h1_parameter[2], fill=None, edgecolor='orange', linestyle='--', label='Hypothesis h1')
        plt.gca().add_patch(h1_rectangle)

    if h2_parameter:
        h2_rectangle = plt.Rectangle((h2_parameter[0], h2_parameter[2]), h2_parameter[1]-h2_parameter[0], h2_parameter[3]-h2_parameter[2], fill=None, edgecolor='purple', linestyle='--', label='Hypothesis h2')
        plt.gca().add_patch(h2_rectangle)

    plt.legend()
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Dataset and Hypotheses')
    plt.grid(True)
    plt.show()

h_true = generate_concept()
x, y, labels = generate_dataset(h_true)
h1_parameter, h1_error = generate_hypothesis_and_error(x, y, labels)
plot_data_and_hypotheses(x, y, labels, h_true, h1_parameter)


x_new, y_new, labels_new = generate_dataset(h_true, n_samples=100)
error_new_data_h1 = calculate_error_on_new_data(x_new, y_new, labels_new, h1_parameter)


print("Empirical error of h1 on new data:", error_new_data_h1)


x_combined = np.concatenate([x, x_new])
y_combined = np.concatenate([y, y_new])
labels_combined = np.concatenate([labels, labels_new])
h2_parameter, h2_error = generate_hypothesis_and_error(x_combined, y_combined, labels_combined)
plot_data_and_hypotheses(x_combined, y_combined, labels_combined, h_true, h1_parameter, h2_parameter)

print("Observations:")
print("Empirical error of h1:", h1_error)
print("Empirical error of h2:", h2_error)
```
